```python
import pandas as pd
numeric_cols=df.select_dtypes(include=['float64','int64']).columns
for col in numeric_cols:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    lower_limit=Q1-1.5*IQR
    upper_limit=Q3+1.5*IQR
    outliers_df=df.loc[(df[col]<lower_limit) | (df[col]>upper_limit)]
    if not outliers_df.empty:
      print(f"outliers found in column {col}")
      outliers_df

```

```python
df_cleaned=df.copy()
import numpy as np
numeric_cols=df.select_dtypes(include=['float64','int64']).columns
for col in numeric_cols:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    lower_limit=Q1-1.5*IQR
    upper_limit=Q3+1.5*IQR
    df_cleaned[col]=np.where(df_cleaned[col]<lower_limit,lower_limit,df_cleaned[col])
    df_cleaned[col]=np.where(df_cleaned[col]>upper_limit,upper_limit,df_cleaned[col])

df_cleaned
```

```python
import pmdarima as pm
import pandas as pd
df_vietnam = df[df['country'] == 'Vietnam'][['temperature_celsius']].copy()
df_vietnam = df_vietnam.sort_index()
df_vietnam = df_vietnam[~df_vietnam.index.duplicated(keep='first')]
train_data = df_vietnam.iloc[:-5]
actual_answers = df_vietnam.iloc[-5:]
model_test = pm.auto_arima(train_data, seasonal=False, stepwise=True)
test_predictions = model_test.predict(n_periods=5)
average_degree_error = abs(actual_answers['temperature_celsius'].values- test_predictions.values).mean()
average_degree_error

```

```python

final_model = pm.auto_arima(df_vietnam, seasonal=False, stepwise=True)
future_forecast = final_model.predict(n_periods=5)
print(future_forecast)
```

```python
import pandas as pd
import numpy as np
import pmdarima as pm
from sklearn.ensemble import IsolationForest
df_vietnam = df[df['country'] == 'Vietnam'][['temperature_celsius']].sort_index()
df_vietnam = df_vietnam[~df_vietnam.index.duplicated(keep='first')]
arima_model = pm.auto_arima(df_vietnam, seasonal=False, stepwise=True)
predicted_temperatures = arima_model.fittedvalues()
df_vietnam['model_error'] = df_vietnam['temperature_celsius'] - predicted_temperatures
X_errors = df_vietnam['model_error'].values.reshape(-1, 1)
iso_forest = IsolationForest(contamination=0.03, random_state=42)
df_vietnam['anomaly_score'] = iso_forest.fit_predict(X_errors)
model_anomalies = df_vietnam[df_vietnam['anomaly_score'] == -1]
model_anomalies
```

```python
from sklearn.inspection import permutation_importance

labels = df_vietnam['anomaly_score']
result = permutation_importance(iso_forest, X_errors, labels, scoring='accuracy')


print(result.importances_mean)
```
